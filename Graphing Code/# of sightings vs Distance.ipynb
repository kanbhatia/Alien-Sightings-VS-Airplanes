```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


file = "Airport_distance.csv"

df = pd.read_csv(file)
df
```

```python
max_distance = df["Distance"].max()
max_distance
```

```python
min_distance = df["Distance"].min()
min_distance
```

```python
# Create bins in which to place values based upon TED Talk views
bins = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
# # Create labels for these bins
group_labels = ["0-5", "5-10", "10-15", "15-20", "20-25", "25-30"
, "30-35", "35-40", "40-45", "45-50"]
```

```python
pd.cut(df["Distance"], bins, labels=group_labels).head()
```

```python
df["Distances"] = pd.cut(df["Distance"], bins, labels=group_labels)
df.head()
```

```python
ted_group = df.groupby("Distances")
print(ted_group["Distance"].count())
ted_group[["lat"]].count()
```

```python
plt.bar(np.arange(len(ted_group["Distance"])), ted_group["lat"].count(), width=0.7, align='center', color="black")
plt.xticks(np.arange(len(ted_group["Distance"])), group_labels)
plt.title("Number of Sightings & Distance from Airport")
plt.xlabel("Distance from Airport (KM)")
plt.ylabel("Number of Sightings")
plt.savefig("numberofsightingsvsdistance.png")
```
