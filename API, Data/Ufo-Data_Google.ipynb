```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests as rp
import gmaps
import requests
import pprint
gkey = ""
import scipy.stats as stats
```

```python
file = "CSV_files\Duplicate-data.csv"
distance_location = "CSV_files\Airport_distance.csv"
df = pd.read_csv(file)
distance_data = pd.read_csv(distance_location)
# df.head()
del distance_data["Unnamed: 0"]
del distance_data["Unnamed: 0.1"]
distance_data.head()
```

```python
df["Airport Name"] = ""
df["Lat Airport"] = ""
df["Lng Airport"] = ""
df.head()
```

```python
params = {
    "radius": 100000,
    "type": "airport",
#     "keyword": "international" or "regional" or "Municipal" or "city",
    "key": gkey
}

                  
for index, row in df.iterrows():
    # get lat, lng from df
    lat = row["lat"]
    lng = row["lng"]

    # change location each iteration while leaving original params in place
    params["location"] = f"{lat},{lng}"

# set up a parameters dictionary


# base url
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    name_address1 = requests.get(base_url, params=params)

    # convert to json
    name_address = name_address1.json()

    try:
        df.loc[index, "Airport Name"] = name_address["results"][0]["name"]
        df.loc[index, "Lat Airport"] = name_address["results"][0]["geometry"]["location"]["lat"]
        df.loc[index, "Lng Airport"] = name_address["results"][0]["geometry"]["location"]["lng"]
    except (KeyError, IndexError):
       
        print("Missing field/result... skipping.")
```
